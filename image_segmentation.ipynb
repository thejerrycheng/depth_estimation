# Webcam Detection

```python
from ultralytics import YOLO
import cv2

# Load a pretrained YOLO model
model = YOLO('yolov8n.pt')  # Make sure 'yolov8n.pt' is the path to your weights file

# Initialize the webcam
cap = cv2.VideoCapture(0)

# Check if the webcam is opened correctly
if not cap.isOpened():
    raise IOError("Cannot open webcam")

while True:
    # Capture frame-by-frame from the webcam
    ret, frame = cap.read()

    # If frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break

    # Perform object detection using the model
    results = model(frame)

    # Retrieve predictions
    preds = results.pred[0]

    # Draw bounding boxes and labels on the frame
    for *box, conf, cls in preds:
        label = model.names[int(cls)]
        box = [round(i, 0) for i in box]  # Convert to integer
        cv2.rectangle(frame, (int(box[0]), int(box[1])), (int(box[2]), int(box[3])), (255, 0, 0), 2)
        cv2.putText(frame, label, (int(box[0]), int(box[1]-10)), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    # Display the resulting frame
    cv2.imshow('YOLO Live Detection', frame)

    # Break the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything is done, release the capture
cap.release()
cv2.destroyAllWindows()
```

# Image Detection

```python
from ultralytics import YOLO
import cv2

# Load a pretrained YOLO model
model = YOLO('yolov8n.pt')  # Make sure 'yolov8n.pt' is the path to your weights file

# Load an image
image_path = 'data_set/my_dataset/photos/person.jpeg'  # Replace with your image path
image = cv2.imread(image_path)

# Check if the image was successfully loaded
if image is None:
    raise ValueError("Image not found")

# Perform object detection using the model
results = model(image)

# Retrieve predictions
preds = results.pred[0]

# Draw bounding boxes and labels on the image
for *box, conf, cls in preds:
    label = model.names[int(cls)]
    box = [round(i, 0) for i in box]  # Convert to integer
    cv2.rectangle(image, (int(box[0]), int(box[1])), (int(box[2]), int(box[3])), (255, 0, 0), 2)
    cv2.putText(image, label, (int(box[0]), int(box[1]-10)), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

# Display the resulting image
cv2.imshow('YOLO Detection', image)
cv2.waitKey(0)  # Wait indefinitely for a key press

# Optionally save the output image with detections
output_path = 'output_image.jpg'  # Replace with the path where you want to save the output image
cv2.imwrite(output_path, image)

# Close the window
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2

# Load a pretrained YOLO model
model = YOLO('yolov8n.pt')  # Make sure 'yolov8n.pt' is the path to your weights file

# Load a video file
video_path = 'data_set/my_dataset/videos/recorded-50521047003.mp4'  # Replace with your video path
cap = cv2.VideoCapture(video_path)

# Check if the video was successfully opened
if not cap.isOpened():
    raise IOError("Cannot open video")

while True:
    # Read the next frame from the video
    ret, frame = cap.read()

    # If frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break

    # Perform object detection using the model
    results = model(frame)

    # Check if the results are in the expected format
    if isinstance(results, list):
        preds = results
    else:
        raise TypeError("The model's output is not in the expected format")

    # Draw bounding boxes and labels on the frame
    for pred in preds:
        for *box, conf, cls in pred:
            label = model.names[int(cls)]
            box = [round(i, 0) for i in box]  # Convert to integer
            cv2.rectangle(frame, (int(box[0]), int(box[1])), (int(box[2]), int(box[3])), (255, 0, 0), 2)
            cv2.putText(frame, label, (int(box[0]), int(box[1]-10)), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    # Display the resulting frame
    cv2.imshow('YOLO Video Detection', frame)

    # Break the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything is done, release the video capture and close all windows
cap.release()
cv2.destroyAllWindows()
```

```python
import torch
from yolov5 import YOLOv5

# Load the model
model = YOLOv5("yolov5s.pt", device="cuda" if torch.cuda.is_available() else "cpu")

# Function to process frames
def process_frame(img):
    # Inference
    results = model(img)
    
    # Results
    results.print()
    return results.render()

# Load a video or webcam
cap = cv2.VideoCapture(0)  # Use 0 for webcam, or replace with video filename

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Convert frame to the format expected by the model
    img = frame[..., ::-1]  # BGR to RGB

    # Process the frame
    detection_frame = process_frame(img)

    # Display the frame
    cv2.imshow('YOLOv5 Detection', detection_frame[..., ::-1])  # RGB to BGR for OpenCV

    # Break loop with 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture
cap.release()
cv2.destroyAllWindows()
```
